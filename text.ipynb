```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Audio-Visual Library Data using NumPy

# Set random seed for reproducibility
np.random.seed(42)

# List of items
items = ['The Great Gatsby', 'Harry Potter', 'Dark Knight',
         'Classical Music Collection', 'Interstellar', 'The Beatles', 'Avengers Endgame', '1984', 'The Lion King']

# Categories (books, movies, audio)
categories = ['Book', 'Movie', 'Audio']

# Random number of checkouts (between 0 to 100 for each item)
checkouts_data = np.random.randint(0, 101, len(items))

# Random number of available copies (between 1 to 5 for each item)
available_copies = np.random.randint(1, 6, len(items))

# Assign categories to items (for simplicity, we'll assign them in a cycle)
item_categories = np.array(categories * 4)[:len(items)]

# Create a Pandas DataFrame
df = pd.DataFrame({
    'Item': items,
    'Category': item_categories,
    'Checkouts': checkouts_data,
    'Available Copies': available_copies
})

# Print the DataFrame to inspect the data
print(df)
```

```python
# 2. Data Analysis

# Calculate the total number of checkouts
total_checkouts = df['Checkouts'].sum()

# Calculate the total number of available copies
total_available = df['Available Copies'].sum()


# Print out statistics
print("\nBasic Library Statistics:")
print(f"Total Checkouts: {total_checkouts}")
print(f"Total Available Copies: {total_available}")
```

```python
# 3. Data Visualization

# Plotting number of checkouts for each item
plt.figure(figsize=(11, 8))

# Bar plot of checkouts for each item
plt.bar(df['Item'], df['Checkouts'], color='Orange')

# Adding titles and labels
plt.title('Checkouts for Each Audio-Visual Item', fontsize=14)
plt.xlabel('Item', fontsize=12)
plt.ylabel('Number of Checkouts', fontsize=12)
plt.xticks(rotation=90)  # Rotate item labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plotting Available Copies vs Checkouts for each category
category_checkouts = df.groupby('Category')['Checkouts'].sum()
category_availability = df.groupby('Category')['Available Copies'].sum()

# Create a bar plot for comparison
plt.figure(figsize=(10,4))

# Plot Checkouts and Available Copies for each category
width = 0.3
x = np.arange(len(category_checkouts))
plt.bar(x - width/2, category_checkouts, width=width, label='Checkouts', color='darkred')
plt.bar(x + width/2, category_availability, width=width, label='Available Copies', color='lightgreen')

# Adding titles and labels
plt.title('Checkouts vs Available Copies by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(x, category_checkouts.index)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
